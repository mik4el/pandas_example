## Making changes to the data

```python
df['Total'] = df['HP']+df['Attack']
```

## Filtering

```python
df.loc[df['Type 1'] == 'Grass']
```

```python
df.loc[~df['Type 1'].str.contains('Grass')]
```

```python
import re

df.loc[df['Type 1'].str.contains('Fire|Grass', regex=True)]
```

## Aggregate Statistics

```python
df.groupby(['Type 1']).mean().sort_values('Attack', ascending=False)
```

```python
df.groupby(['Type 1']).sum().sort_values('Legendary', ascending=False)
```

```python
df['count'] = 1
df.groupby(['Type 1', 'Type 2']).count()['count']
```

## Graphing

```python
attacks = df.groupby(['Type 1']).mean().sort_values('Attack', ascending=False)["Attack"]
```

```python
from matplotlib import pyplot as plt
```

```python
attack_plot = attacks.plot(kind='bar')
attack_plot.set_xlabel("Type 1 Pokemon")
attack_plot.set_ylabel("Attack (mean)")
```
